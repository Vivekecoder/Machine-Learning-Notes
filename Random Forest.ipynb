```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
x=pd.DataFrame(data['data'],columns=list(data['feature_names']))
```

```python
y=data['target']
y=pd.Series(y)
```

```python
y=y.map({0:'Setosa',1:'Versicolor',2:'Virginica'})
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=10)
```

```python
model.fit(x,y)
```

```python
x.tail()
```

```python
y.tail()
```

```python
model.estimators_
```

```python
model.predict([[5.9,3.0,5.1,1.8]])
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import plot_tree
```

```python
data['feature_names']
```

```python
list(data['target_names'])
```

```python
plt.figure(figsize=(10,10))
for i in range(10):
    plot_tree(model.estimators_[i],filled=True,class_names=list(data['target_names']),feature_names=data['feature_names'])
    plt.savefig(f'Ranfom_forest{i}.jpg')
```
